### Encoding

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
# Using LabelEncoder in my target variable
label = LabelEncoder().fit_transform(df["y"])
```

```python
label
```

```python
df["y"] = label
```

```python
df["y"].head()
```

```python
df.marital.replace({'married':3, 'single':2, 'divorced':1, 'unknown':0},inplace=True)
df.default.replace({'no':1, 'unknown':0, 'yes':2},inplace=True)
df.housing.replace({'no':1, 'yes':2, 'unknown':0},inplace=True)
df.loan.replace({'no':1, 'yes':2, 'unknown':0},inplace=True)
df.contact.replace({'telephone':0, 'cellular':1},inplace=True)
df.month.replace({'may':5, 'jun':6, 'jul':7, 'aug':8, 'oct':10, 'nov':11, 'dec':12, 'mar':3, 'apr':4,  'sep':9},inplace=True)
df.day_of_week.replace({'mon':1, 'tue':2, 'wed':3, 'thu':4, 'fri':5},inplace=True)
df.poutcome.replace({'nonexistent':0, 'failure':1, 'success':2},inplace=True)
df.y.replace({'no':0, 'yes':1},inplace=True)
```

```python

# Identify categorical columns that need label encoding
categorical_columns = df.select_dtypes(include='object').columns

# Apply label encoding to categorical columns
label_encoder = LabelEncoder()
df[categorical_columns] = df[categorical_columns].apply(lambda x: label_encoder.fit_transform(x.astype(str)))
```
